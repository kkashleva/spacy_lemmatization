```
#загружаем модель для русского языка
import spacy
nlp = spacy.load("ru_core_news_sm")

#в кавычках указано название модели. https://spacy.io/usage/models - здесь можно выбрать модели для других языков
```

# Токенизация, POS-tagging и определение начальной формы слова - лемматизация

```
#подключаемся к гугл-диску

from google.colab import drive
drive.mount('/content/drive')
```

```
# открываем файл, меняем название файла при необходимости

with open('/content/drive/MyDrive/bulgakov.txt', 'r', encoding='utf8') as f:
  text = f.read()
```

```
# смотрим, что файл считался (отображаются первые 20 символов)
text[:20]
```

```
# если не нужны части речи
doc = nlp(text)

lemmatized = []

# удаляем мусор и стоп-слова
for token in doc:
  if (token.is_stop != True) and (token.is_punct != True) and (token.is_space != True) and (token.is_digit != True):
    lemmatized.append(token.lemma_)
```

```
# смотрим первые 10 слов
lemmatized[:10]
```

```
# записываем леммы в файл и сохраняем на свой гугл-диск

with open('/content/drive/MyDrive/lemmatized.txt', 'w', encoding='utf8') as file:
  file.writelines(f"{item} " for item in lemmatized)
```
